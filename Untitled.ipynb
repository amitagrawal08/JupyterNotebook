```python
metrocities = ['Mumbai','New Delhi','Bangalore','Chennai','Hyderabad','Kolkata','Pune','Ahmedabad']
tier1cities = ['Agra','Nagpur','Indore','Jaipur','Mangalore']

def CityClassifier(input):
    city=input[1]
    if city in metrocities:
        return "It is under Metro City"
    elif city in tier1cities:
        return "It is under Tier1 city"
    else:
        return "It is under Tier2 and other cities"
```
